```python
import pandas as pd
from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import copy

# Convertir las fechas a objetos datetime
dates = [datetime.strptime(row[1], "%d/%m/%y %H:%M") for row in data]
states = [row[2] for row in data]

# Agrupar las fechas por día y calcular los intervalos en horas entre días consecutivos
grouped_dates = {}
for date, state in zip(dates, states):
    day = date.date()
    if day not in grouped_dates:
        grouped_dates[day] = []
    grouped_dates[day].append((date, state))
grouped_dates_state = copy.deepcopy(grouped_dates)
# Ordenar las fechas dentro de cada día
for day, day_data in grouped_dates.items():
    day_dates = [x[0] for x in day_data]
    grouped_dates[day] = sorted(day_dates)

# Calcular los intervalos en horas entre días consecutivos
intervals = {}
for day, day_dates in grouped_dates.items():
    last_occurrence = day_dates[-1]
    next_day = day + timedelta(days=1)
    if next_day in grouped_dates:
        next_occurrence = grouped_dates[next_day][0]
        interval = next_occurrence - last_occurrence
        intervals[day] = interval.total_seconds() / 3600

# Calcular la cantidad de días sin 'Húmedo'
humid_intervals = {}
last_humid_date = None
for day, day_data in grouped_dates_state.items():
    for date, state in day_data:
        if state == 'Húmedo':
            if last_humid_date is not None:
                humid_intervals[last_humid_date] = (day - last_humid_date).days
            last_humid_date = day
            break

# Crear un DataFrame con las características y el objetivo
df = pd.DataFrame({
    'Día': list(intervals.keys()),
    'Intervalo en horas': list(intervals.values())
})
df['Días sin Húmedo'] = df['Día'].map(humid_intervals)
df['Objetivo'] = (df['Días sin Húmedo'] >= 30) & (df['Intervalo en horas'] >= 10)

df = df.dropna()

print(df)

# Dividir el conjunto de datos en entrenamiento y prueba
X = df[['Intervalo en horas', 'Días sin Húmedo']]
y = df['Objetivo']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=10)

# Entrenar un modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluar el modelo
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Error cuadrático medio:', mse)

# Predecir cuándo se alcanzarán los umbrales especificados
print(model.intercept_)
print(model.coef_[1])
print(model.coef_)
#predicted_date = datetime.now().date() + timedelta(days=int((30 - model.intercept_) / model.coef_[1]))
#print('Fecha estimada para alcanzar los umbrales:', predicted_date)
```
