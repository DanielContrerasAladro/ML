# Ejemplos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y,'red')
plt.title('grafico')
plt.xlabel('a')
plt.ylabel('a2')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Programación orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.2,0.8,0.8])
grafico.plot(x,y,'g')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')
grafico2 = figura.add_axes([0.2,0.4,0.3,0.3])
grafico2.plot(y,x,'b')
```

# Multigráficos

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)
graficos[0][0].plot(x,y,'b')
graficos[0][1].plot(x,y,'r')
graficos[1][0].plot(x,y,'y')
graficos[1][1].plot(x,y,'g')

graficos[0][0].set_title('b')
graficos[0][1].set_title('r')
graficos[1][0].set_title('y')
graficos[1][1].set_title('g')
```

# Tamaños

```python
figura = plt.figure(figsize=(8,1))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

# Dos gráficos en una figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.8,0.8])
graficos.plot(x,x**2,label='cuadrado')
graficos.plot(x,x**3,label='cubo')
graficos.legend(loc=0)
```

# Estilos

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red',linewidth=10,alpha=0.4,linestyle=':')
```

# Marcadores

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'b',marker='s',markersize=5)
```
