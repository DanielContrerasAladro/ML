```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('vehiculos.csv')
```

```python
dataset.head()
```

```python
dataset.vehicle_class.unique()
```

```python
y = dataset.vehicle_class
x = dataset.drop('vehicle_class',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=45)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
y_pred = knn.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
tasa_error = []
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    i_predict = knn.predict(X_test.values)
    tasa_error.append(np.mean(i_predict != y_test))
tasa_error
```

```python
valores = range(1,30)
plt.plot(valores, tasa_error, color='g', marker='o', markerfacecolor='r', markersize=5) 
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test.values)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
